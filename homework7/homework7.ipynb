```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    y =  A*np.sin(w*x)
    return y
print(make_sine_wave(3,5,7))
```

```python
x = np.linspace(0, 2*np.pi, 1000)
```

```python
amplitude = [0.5, 1, 1.5, 2, 2.5]
```

```python
frequency = [1, 2, 3, 4, 5]
```

```python
fig,ax = plt.subplots(1,3,figsize=(25,5))
```

```python
fig, ax = plt.subplots(1, 3, figsize=(25, 5))

for A, w in zip(amplitude, frequency):
    ax[0].plot(x, make_sine_wave(x, A, 1), label=f"A={A}")
    ax[1].plot(x, make_sine_wave(x, 1, w), label=f"f={w}")
    ax[2].plot(x, make_sine_wave(x, A, w), label=f"A={A}, f={w}")

ax[0].set_title("Varying Amplitude")
ax[1].set_title("Varying Frequency")
ax[2].set_title("Varying Both Amplitude and Frequency")

for a in ax:
    a.legend()
    a.set_xlabel("x")
    a.set_ylabel("y")

plt.show()
```
