```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("/Users/sophiachoi/python_decal_fall25/course_assignments/homework7/star_data.txt", skiprows = 1)
```

```python
np.shape(data)
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]
```

```python
np.min(magnitude)
```

```python
np.max(magnitude)
```

```python
np.min(temperature)
```

```python
np.argmin(temperature)
#star 1923
```

```python
np.max(temperature)
```

```python
np.argmax(temperature)
#star 2613
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)
```

```python
plt.figure(figsize=(8,6))
```

```python
plt.scatter(log_temperature, log_magnitude, label = "stars")
plt.scatter(log_temperature[np.argmin(magnitude)], log_magnitude[np.argmin(magnitude)],
            color="black", marker="*", s=200, label="Brightest")
plt.scatter(log_temperature[np.argmax(magnitude)], log_magnitude[np.argmax(magnitude)],
            color="purple", marker="*", s=200, label="Dimmest")

plt.scatter(log_temperature[np.argmin(temperature)], log_magnitude[np.argmin(temperature)],
            color="green", marker="*", s=200, label="Hottest")

plt.scatter(log_temperature[np.argmax(temperature)], log_magnitude[np.argmax(temperature)],
            color="orange", marker="*", s=200, label="Coldest")
plt.gca().invert_xaxis() 
plt.gca().invert_yaxis()

plt.legend()
plt.title("HertzsprungRussell Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")

plt.show()
```
