```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = sns.load_dataset("titanic")
```

```python
df = dataset.dropna(axis=1)
```

```python
df.shape
#891 rows, 11 columns
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns
#survived numerically shows whether the person lived (1 for yes 0 for no)
#pclass shows the class on the ship they were in
#sex is gender
#sibsp is the number of siblings/spouses
#parch is the number of parents or children aboard
#fare is how much they payed for their ticket
#class is the class of the ship but in words instead of numbers
#who lists the gender again, but in "man" and "woman"
#adult_male shows the gender(again) but in booleans
#alive shows whether they survived but with yes and no
#alone shows whether they were alone with booleans
```

```python
fig, ax = plt.subplots(2,2, figsize=(10,10))
ax[0,0].scatter(df["pclass"],df["fare"], color="blue", alpha=0.7)
ax[0,0].set_title("Pclass vs Fare")
ax[0,0].set_xlabel("Passenger Class (1 = First, 2 = Second, 3 = Third)")
ax[0,0].set_ylabel("Fare (USD)")

sns.scatterplot(x="sibsp", y="parch", hue="survived", data=df, ax=ax[0,1])
ax[0,1].set_title("Siblings/Spouses vs Parents/Children")
ax[0,1].set_xlabel("Siblings/Spouses Aboard (people)")
ax[0,1].set_ylabel("Parents/Children Aboard (people)")
ax[0,1].legend(title="Survived", loc="upper right")

sns.histplot(df["fare"], ax=ax[1,0], color="green")
ax[1,0].set_title("Fare Distribution")
ax[1,0].set_xlabel("Fare (USD)")
ax[1,0].set_ylabel("Passengers (people)")

sns.boxplot(df["pclass"], ax = ax[1,1], color="purple")
ax[1,1].set_title("Boxplot of Passenger Class")
ax[1,1].set_xlabel("Passengers")
ax[1,1].set_ylabel("Class (1 = first, 2 = second, 3 = third)")
plt.show()
```

```python
#I didn't have a lot of numerical data to work with, but I knew that the fare was the most scattered form of data. The first plot shows the price of the tickets by passenger class, the second shows the numebr of parents and children in comparison to the number of siblings and spouses with extra coloring by whether the person survived or not. The third shows the fare distribution based on the total number of passengers on the ship, and the last shows a boxplot of the passanger class also based on the total number of passangers (it looks weird because there were only 3 classes)
#I learned that the price of a ticket for second and third class seems to be relatively similar, which doesn't make a lot of sense. Otherwise, there isn't too much correlation based on the data that I chose, but again there was not a lot to work with.
```
